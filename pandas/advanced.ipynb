#### **Correlation**

Clean data
- fill null value with mean value

```python
import pandas as pd

df = pd.read_csv('data.csv')

# fill null calories with mean calories
mean_calories = df['Calories'].mean()
df.fillna({'Calories' : mean_calories}, inplace=True)
```

Calculate correlation between column

```python
df.corr()
```

#### **Plotting**

Plot scatter graph

```python
import matplotlib.pyplot as plt

df.plot()
plt.show()
```

```python
df.plot(kind='scatter', x='Duration', y='Calories')
plt.show()
```

```python
df.plot(kind='scatter', x='Duration', y='Maxpulse')
plt.show()
```

Plot histogram graph

```python
df['Duration'].plot(kind='hist')
plt.show()
```
