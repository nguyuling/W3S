Import pandas library

```python
import pandas as pd
```

#### **Dictionary -> DataFrame (2d array)**

```python
car = {
    'cars' : ['BMW', 'Volvo', 'Ford'],
    'passings' : [3, 7, 2]
}

df_car = pd.DataFrame(car)

df_car
```

Checking pandas version

```python
print(pd.__version__)
```

#### **List -> Pandas Series (1d array)**

```python
a = [1, 7, 2]

series_a = pd.Series(a)

series_a
```

create label

```python
series_a = pd.Series(a, index=['x', 'y', 'z'])

series_a
```

Acess item in a Pandas Series

```python
print(series_a['x'])
```

#### **Dictionary -> Pandas Series**

```python
calories = {
    'day1' : 420,
    'day2' : 380,
    'day3' : 390
}

# select only day1 and day2 by using index
series_calories = pd.Series(calories, index=['day1', 'day2'])
series_calories
```

#### **DataFrames**

```python
data = {
    'calories' : [420, 380, 390],
    'duration' : [50, 40, 45]
}

df_data = pd.DataFrame(data)

df_data
```

Locate row to retrieve row(s) attributes

```python
df_data.loc[0]
```

```python
df_data.loc[[0, 1]]
```

Renamed labels

```python
df_data = pd.DataFrame(data, index=['day1', 'day2', 'day3'])

df_data
```

Locate row using renamed label

```python
df_data.loc['day2']
```

Locate row using int index

```python
df_data.iloc[1]
```
