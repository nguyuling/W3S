### **Linear Regression**
- scatter plot + linear line showing the relationship of 2 variables
- r value between -1 and 1, 0 means no relationship

Scatter plot + linear regression

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept
    # y = mx + c

mymodel = list(map(myfunc, x))
# run each value of the x array through the function to result in new array with new values for the y-axis

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show
```

Relationship
- ranges from -1 to 1
- 0 means no relationship
- 1 means directly proportional
- -1 means inversely proportional 

```python
print(r)
# -0.76 shows that there is a remarkable relationship, however not perfect
```

Predict future value
- pass the x value to myFucn to predict the y value

```python
print(myfunc(10))
```

### **Polynomial Regression**
- scatter plot + polynomial line (deg=3) between 2 variables (x & y)
- r-squared value between -1 and 1, 0 means no relationship

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))
# polynomial model, 3 is the degree of polynormial, the higher the more curvy

myline = np.linspace(1, 22, 100)
# the line will be display from 1 to 22

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

R-squared
- value from -1 to 1 showing the strength of the relationship

```python
from sklearn.metrics import r2_score

print(r2_score(y, mymodel(x)))
```

Example of a bad fit

- where the polynomial line could hardly intercept any dots

```python
import numpy as np
import matplotlib.pyplot as plt

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(0, 100, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

from sklearn.metrics import r2_score
print(r2_score(y, mymodel(x)))
```

### **Multiple Regression**

### **Scaling**

Scale the X values using (x-u)/s
- x = original value
- u = mean
- s = standard deviation

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('data.csv')
X = df[['Weight', 'Volume']]
y = df['CO2']

scaledX = StandardScaler().fit_transform(X)
print(scaledX)
```

Predict y value

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

regr = linear_model.LinearRegression()
regr.fit(scaledX, y) # pass the scaled X

scaled = StandardScaler().fit_transform([[2300, 1.3]]) # transform the X values

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)
```

### **Train/Test**
Split dataset into train/test to see how fitting any model is, on the dataset

Generate a random dataset using numpy

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

Split into
- train (0-80)
- test (80-100)

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

Fit the train dataset using polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt

mymodel = np.poly1d(np.polyfit(train_x, train_y, 4)) # poly regr of degree 4
myline = np.linspace(0, 6, 100) # x-axis 0 to 6

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

R-squared of train/test set
- indicating how well the data is fitting the model

```python
from sklearn.metrics import r2_score

train = r2_score(train_y, mymodel(train_x))
test = r2_score(test_y, mymodel(test_x))

print(train)
print(test)
```

Use the model to predict y value

```python
print(mymodel(5))
```

### **Decision Tree**
- data must be numerical

Convert the `feature` and `target` data to numerical using `.map()` from pandas

```python
import pandas as pd

df = pd.read_csv('data2.csv')

d = {'UK':0, 'USA':1, 'N':2} # dictionary
df['Nationality'] = df['Nationality'].map(d)

d = {'YES':1, 'NO':0}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.read_csv('data2.csv')

#! convert data to numerical
d = {'UK' : 0, 'USA' : 1, 'N' : 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES' : 1, 'NO' : 0}
df['Go'] = df['Go'].map(d)

#! setup x-features (all columns other than the target column) & y-target
features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']

#! create decision tree
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features) 
```

Predict values

```python
print(dtree.predict([[40, 10, 7, 1]]))
print(dtree.predict([[40, 10, 6, 1]]))
```
