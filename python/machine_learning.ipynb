### **Statistics**
- mean : `np.`**`mean`**`([list])`

- median : `np.`**`median`**`([list])`

- mode : `stats.`**`mode`**`([list])`

- standard deviation : `np.`**`std`**`([list])`

- variance : `np.`**`var`**`([list])`

- percentile : `np.pecentile([list], 0-100)`


```python
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

Mean

```python
import numpy as np
print(np.mean(speed))
```

Median

```python
print(np.median(speed))
np.median(speed)
```

Mode

```python
from scipy import stats
stats.mode(speed)
```

Standard deviation

```python
print(np.std(speed))
```

Variance

```python
print(np.var(speed))
```

Percentile

```python
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

print(np.percentile(ages, 75))
print(np.percentile(ages, 90))
```

### **Random Uniform Data Distribution**

`random.`**`uniform`**`(min, max, size)`
- genereta a random dataset
- then store it as a list / tupple

`plt.`**`hist`**`([list], n)`
- plot the random dataset as a histogram with n bar

```python
from numpy import random
import matplotlib.pyplot as plt

myList = random.uniform(0.0, 0.5, 100000)
plt.hist(myList, 1000)
plt.show()
```

### **Random Normal Data Distribution**

`random.`**`normal`**`(mean, std, size)`
- genereta a normal distributed random dataset
- then save it as a list / tuple

`plt.`**`hist`**`([list], n)`
- same as above

```python
from numpy import random
import matplotlib.pyplot as plt

myList = random.normal(5.0, 1.0, 100000)

plt.hist(myList, 1000)
plt.show()
```

### **Scatter Plot**

`plt.`**`scatter`**`(x, y)`
- first, obtain/generate a list of dataset (uniform/normal) for x- and y-axis with equal size

- plot a scatter plot using the axes

Age of car VS. Speed of car

```python
import matplotlib.pyplot as plt
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(age, speed)
plt.show()
```

Random Normal Data Distribution

```python
from numpy import random
import matplotlib.pyplot as plt

# generate x and y axis data with (mean, std, equal-size)
x = random.normal(5.0, 1.0, 1000)
y = random.normal(10.0, 2.0, 1000)

plt.scatter(x, y)
plt.show()
```

### **Linear Regression**
(see example below)

Scatter plot + linear regression

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myFunc(x):
    return slope * x + intercept
    # y = mx + c

myModel = list(map(myFunc, x))
# run each value of the x array through the function to result in new array with new values for the y-axis

plt.scatter(age, speed)
plt.plot(x, myModel)
plt.show
```

Relationship
- ranges from -1 to 1
- 0 means no relationship
- 1 means directly proportional
- -1 means inversely proportional 

```python
print(r)
# -0.76 shows that there is a remarkable relationship, however not perfect
```

Predict future value
- pass the x value to myFucn to predict the y value

```python
print(myFunc(10))
```
