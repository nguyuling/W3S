#### **Unpacking a collecion**

```python
fruits = ['apple', 'banana', 'cherry'] # list
x, y, z = fruits

print(x, y, z)
```

#### **Global variable**

```python
x = 'apple'
y = 'apple'

def myfunc():
    global x
    x = 'banana'
    y = 'banana'

myfunc()
print(x, y)
# x is changed in the function by calling global
# y is not changed bc global has not been called
```
